```python
import sys
sys.path.append(('../Helper_Scripts'))
from data_retrieval import data
import matplotlib.pyplot as plt
import contextily as cx
```

```python
data = data(
    data_base_path = "..\Data",
    census_api_key = "APIKeyHere",
    county_code = "073"
)
```

```python
fig, ax = plt.subplots(
    nrows = 1,
    ncols = 1,
    figsize = (16, 12)
)

#Plotting routes
data.routes_df.plot(ax = ax, linewidth = 0.2, alpha = 1, color = "blue")

#Plotting an OSM basemap
cx.add_basemap(
    ax = ax,
    source = cx.providers.OpenStreetMap.Mapnik,
    zoom = 11,
    crs = data.routes_df.crs,
)

#Plotting the boundaries of the census tracts
data.tract_df.boundary.plot(ax = ax, alpha = 1, color = "black", linewidth = 0.1)
data.tract_df.plot(ax = ax, alpha = 0.5, column = "total_route_length_to_area_ratio", cmap = "hot_r")

ax.axis("off")
plt.title("MTS Routes in SD County", fontsize = 25)
```
