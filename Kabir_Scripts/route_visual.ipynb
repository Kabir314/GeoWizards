```python
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
path = '..\\Data\\transit_routes_datasd\\transit_routes_datasd.shp'
transit_routes = gpd.read_file(path).to_crs("EPSG:4326")

path = "..\\Data\\tl_2021_06_tract\\tl_2021_06_tract.shp"
tract_df = gpd.read_file(path).to_crs("EPSG:4326")
tract_df = tract_df.loc[tract_df['COUNTYFP']== '073',]
```

```python
fig,ax = plt.subplots(1,1,sharex=True,sharey=True,figsize=(16,12))
tract_df.boundary.plot(ax=ax,linewidth=0.1,color='black')
transit_routes.plot(ax=ax,
                    linewidth = 0.1,
                    color='red')
plt.title("MTS Routes",fontsize=25)
plt.savefig('transit_routes.png')
```
